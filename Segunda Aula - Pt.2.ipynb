```python
# Esta célula importa os pacotes que iremos utilizar
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
# O comando abaixo é utilizado para que a figura seja mostrada no próprio notebook
%matplotlib inline 
```

```python
# Condições Inicias
v0 = 0
x0 = 0.1
# Parametros dados e timestep
wn = 2*2*sp.pi
wf = 2.1*2*sp.pi
q = 0.01
f0 = 1
r = sp.linspace(0,2,100)
```

```python
def g(q):
    y = (f0)/((((1-r**2)**2 + (2*q*r)**2))**(1/2))
    plt.plot(r,y)
```

```python
interact(g, q=(0,1,0.01))
```

```python
def theta(q):
    z=sp.arctan((2*q*r/wn)/(1-r**2))
    plt.plot(r,z)
```

```python
interact(theta, q=(0,2*sp.pi,0.01))
```
